```python
import tkinter as tk
from tkinter import messagebox


   
def order1():
    #Output window 
    window3 = tk.Tk(window1)
    window3.title('Delivery Fee')
    window3.geometry('500x500')
            
    label_1 = tk.Label(window3, text = "Your delivery fee is N5,000")
    label_1.pack()
    
    window1.mainloop()

def order2():
    #Output window 
    window3 = tk.Toplevel(window1)
    window3.title('Delivery Fee')
    window3.geometry('500x500')
            
    label_1 = tk.Label(window3, text = "Your delivery fee is N3,500")
    label_1.pack()
    window1.mainloop()

def order3():
    #Output window 
    window3 = tk.Toplevel(window1)
    window3.title('Delivery Fee')
    window3.geometry('500x500')
            
    label_1 = tk.Label(window3, text = "Your delivery fee is N10,000")
    label_1.pack()
    window1.mainloop()
    
       
def order4():
    #Output window 
    window3 = tk.Toplevel(window1)
    window3.title('Delivery Fee')
    window3.geometry('500x500')
            
    label_1 = tk.Label(window3, text = "Your delivery fee is N5,000")
    label_1.pack()
    window1.mainloop()


def submit():
    location = location_entry.get()
    weight = float(weight_entry.get())
    
    if location == 'Ibeju-Lekki'  and weight >= 10.0:
        order1()

    elif location == 'Ibeju-Lekki'  and weight < 10.0:
        order2()

    elif location == 'Epe'  and weight >= 10.0:
        order3()

    elif location == 'Epe'  and weight < 10.0:
        order4()
    
    else:
        messagebox.showerror('Delivery Failed','Invalid location or weight')


#Create the intro window
window1 = tk.Tk()
window1.title('Welcome to Simi Services')
window1.geometry('500x500')

#Create location label and entry
location = tk.Label(window1, text = "location: ")
location.pack()
location_entry = tk.Entry(window1)
location_entry.pack()

#Create password label and entry
weight = tk.Label(window1, text = "Weight: ")
weight.pack()
weight_entry = tk.Entry(window1)
weight_entry.pack()

#Create submit button
submit_button = tk.Button(window1,text="Submit", command = submit)
submit_button.pack()

window1.mainloop()
```
